```python
# importing libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# loading dataset to pandas dataframe.
df = pd.read_csv('./Car_sales - Car_sales.csv')
df.head()
```

```python
# cheking the count of null values.
df.isnull().sum()
```

```python
# finding % of null values.
df.isnull().mean()*100
```

```python
# Drop missing values
df = df.dropna()

# Drop duplicates
df = df.drop_duplicates()
```

```python
# after removing null values, again checking null values.
df.isnull().sum()
```

```python
# finding the data types of dataset's features.
df.dtypes
```

```python
# Counting the no. of each manufacturer type
manufacturers_counts = df['Manufacturer'].value_counts()

# Visualizing the bar graph
plt.figure(figsize=(10, 6))
manufacturers_counts.plot(kind='bar')
plt.xlabel('Manufacturer Type')
plt.ylabel('Count of Manufacturer Type')
plt.title('Distribution of Manufacturer Type')
plt.show()
```

```python
# finding the no. of each vehicle type
vehicle_counts = df['Vehicle_type'].value_counts()

# drawing the pie chart
plt.figure(figsize=(8, 8))
vehicle_counts.plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.title('Type of Vehicles')
plt.show()
```

```python
# Calculating the average price
average_price = df['Price_in_thousands'].mean()
average_price
```

```python
# chekcing the information about dataset. e.g datatypes, null-values, no. of features etc.
df.info()
```

```python
# Calculate variance and standard deviation
numerical_columns = ['Sales_in_thousands', '__year_resale_value', 'Price_in_thousands', 'Engine_size', 
                     'Horsepower','Wheelbase','Width','Length','Curb_weight','Fuel_capacity','Fuel_efficiency',
                     'Power_perf_factor']
variance = df[numerical_columns].var()
std_deviation = df[numerical_columns].std()
print("------------ Variance ------------")
print(variance)
print("\n------ Standard Deviation ------")
print(std_deviation)
```

```python
# Calculatint the  IQR and identifyint the outliers
outliers = {}
for column in numerical_columns:
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    column_outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    outliers[column] = column_outliers

outliers
```

```python
# Importing the necessary libraries
from scipy.stats import skew

# Ploting histograms and calculate skewness
for column in numerical_columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[column], kde=True)
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {column}')
    plt.show()
    skewness = skew(df[column])
    print(f'Skewness of {column}: {skewness:.2f}')
```
